Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt

plt.figure(figsize=(9.5,6.5))
plt.xlabel('X Label Axis')
plt.ylabel('Y Label Axis')
plt.title('Grafik Cos Sin')
plt.plot(x,cos,color='r',label='Cos x')
plt.plot(x,cos3,ls='dashed',color='g',label='3 Cosx')
plt.plot(x,sin,ls='dashdot',color='b',label='Sin(x-90)')
plt.plot(x,sin2,ls='none',marker='x',color='y',label='2 Sinx')
plt.legend(loc='center')
plt.show()
```
